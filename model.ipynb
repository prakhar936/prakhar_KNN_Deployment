```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split as tt
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
import pickle
```

```python
dtrain = pd.read_csv(r"C:\Users\91936\Desktop\Logistic and KNN\mobile_data.csv")
dtrain.drop(['blue','pc' ,'clock_speed','dual_sim','fc','four_g','int_memory','m_dep','mobile_wt','n_cores','sc_h','sc_w','talk_time','three_g','touch_screen','wifi'], axis = 1, inplace = True)
```

```python
x = dtrain.drop('price_range',axis = 1)
y = dtrain['price_range']

x_axis = []
y_axis = []
for i in range(1,20,2):
    k = KNeighborsClassifier(n_neighbors = i)
    score = cross_val_score(k,x,y)
    
    x_axis.append(i)
    y_axis.append(score.mean())
    
plt.plot(x_axis,y_axis)
plt.show()
```

```python
algo = KNeighborsClassifier(n_neighbors = 9)
algo.fit(x,y)
```
